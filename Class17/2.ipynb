Create basic database in sqllite using "circuits.csv" file

```julia
using SQLite

# Create a new SQLite database
db = SQLite.DB("circuits.db")

# Create a table for "circuits"
SQLite.execute(db, """
CREATE TABLE IF NOT EXISTS circuits (
    circuitsId INTEGER,
    circuitRef TEXT,
    name TEXT,
    location TEXT,
    country TEXT,
    lat REAL,
    Lng REAL,
    alt REAL,
    url TEXT
);
""")

# Send message to confirm table creation
println("Table 'circuits' created successfully.")
```

Import DataFrames and CSV packages

```julia
# Import DataFrame, CSV
import Pkg
Pkg.add("DataFrames")
Pkg.add("CSV")
```

import data from circuits.csv to circuits.db (SQLite)

```julia
using CSV
using DataFrames
using SQLite

# Read the CSV file into a DataFrame
df = CSV.read("circuits.csv", DataFrame)

# Insert data in table in SQLite database
for row in eachrow(df)
    try
        SQLite.execute(db, "INSERT INTO circuits VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)", 
            Tuple(row))
    catch e
        println("Error inserting row: ", row)
        println("Error message: ", e)
    end
end

# Message confirm correct inserte data in table in SQLite database
println("Data inserted successfully into 'circuits' table.")
```

Consultar para agrupar y contar el numero de circuitos en cada pais

```julia
using SQLite, DataFrames, PrettyTables

db = SQLite.DB("circuits.db")

query = DBInterface.execute(db, """
SELECT country, COUNT(*) AS total_circuits
FROM circuits
GROUP BY country
ORDER BY total_circuits DESC;
""")

df = DataFrame(query)
pretty_table(df)
SQLite.close(db)
```

Los circuitos con mayor altitud

consultar para listar los cinco circuitos con mayor altitud

```julia
# Import the necessary packages
using SQLite, DataFrames, PrettyTables

# Create connection to the SQLite database
db = SQLite.DB("circuits.db")

# Query the database to get 5 highest altitudes of circuits.
query = DBInterface.execute(db, """
SELECT name, country, alt
FROM circuits
ORDER BY alt DESC;
LIMIT 5;
""")

# Convert the query result to a DataFrame
df = DataFrame(query)

# Display the DataFrame using PrettyTables
pretty_table(df)

# Close the database connection
SQLite.close(db)
```

# SubQueris

Conjunto anidados de SELECT

```julia
# Import the necessary packages
using SQLite, DataFrames, PrettyTables

# Create connection to the SQLite database
db = SQLite.DB("circuits.db")

# Query the database.
query = DBInterface.execute(db, """
SELECT name, country, alt
FROM circuits
WHERE country IN (
    SELECT country
    FROM circuits
    GROUP BY country
    HAVING AVG(alt) > 5.0
)
ORDER BY country, alt DESC;
""")

# Convert the query result to a DataFrame
df = DataFrame(query)

# Display the DataFrame using PrettyTables
pretty_table(df)

# Close connection to the database
SQLite.close(db)
```
